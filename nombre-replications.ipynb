```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
nEchantillons = 50
plt.figure(figsize = (8,4))
for j in range(50):
    y = norm().rvs(nEchantillons)
    x = list(range(1,nEchantillons))
    plt.plot(x,[np.mean(y[:j]) for j in x], 'b', alpha = 0.2)
plt.xlabel('Taille de l\'échantillon')
plt.ylabel('Moyenne de l\'échantillon')
plt.savefig('nombre-replications-gaussienne.pdf')
```
