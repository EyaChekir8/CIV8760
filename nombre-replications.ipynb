```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
nEchantillons = 50
for i in range(4):
    plt.figure()
    #plt.subplot(2, 2, i+1)
    for j in range(10):
        x = norm().rvs(nEchantillons)
        plt.plot(list(range(1,nEchantillons)),[np.mean(x[:j]) for j in range(1,nEchantillons)])
#plt.savefig('nombre-replications-gaussienne.pdf')
```
